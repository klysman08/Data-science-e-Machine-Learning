```
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

Carregando imagem

```
img = cv2.imread("limi.png", 0) # 0: carregando em tons de cinza
```

##Limiarização

Função

```
limiar, imgLimiar = cv2.threshold(img, 150, 255, cv2.THRESH_BINARY) 
# <= do valor estabelecido (150) o pixel recebe 0 e acima disso recebe 255 referente ao branco 
```

```
cv2_imshow(imgLimiar)
```

##Limiarização Adaptativa

```
img = cv2.imread("moedas.jpg", 0)
```

```
adp1 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, \
                             cv2.THRESH_BINARY, 11, 2) #tamanho da visiança deve ser impar. quanto maior mais informaçaões, quanto menor, a media será local.
```

```
cv2_imshow(img)
print("Limiarização")
cv2_imshow(adp1)
```
