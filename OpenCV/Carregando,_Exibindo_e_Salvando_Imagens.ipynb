```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

##Carregando uma imagem e exibindo usando cv2_imshow
Separando os canais de cores utilizando a função split

```
img = cv2.imread("klysman.jpg")


b,g,r = cv2.split(img)
img_BRG = cv2.merge((b,g,r))

cv2_imshow(img_BRG)

mzero=np.zeros(b.shape,np.uint8)
img_B = cv2.merge((b,mzero,mzero))
cv2_imshow(img_B)
print (img_B.shape)

img_G = cv2.merge((mzero,g,mzero))
cv2_imshow(img_G)
print(img_G.shape)

img_R = cv2.merge((mzero,mzero,r))
cv2_imshow(img_R)

img_RG = cv2.merge((mzero,g,r))
cv2_imshow(img_RG)

```
