```
import numpy as np 
import cv2
from matplotlib import pyplot as plt 
```

##Criando função para plotar imagens com o matplotlib

```
def showImage(img):
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  plt.imshow(img)
  plt.show()
```

##Função Principal

```
obj_img = cv2.imread("klysman.jpg")
altura, largura, canais = obj_img.shape

print("Dimensões da imagem: " + str(largura) + "x" + str(altura))
print("Canais de cores: ", canais)

showImage(obj_img)
```

##Alterando os canais de cores por FOR

```
def getColor(img, x , y):
  return img.item(y,x,0), img.item(y,x,1), img.item(y,x,2)

def setColor(img,x,y,b,g,r):
  img.itemset((y,x,0),b)
  img.itemset((y,x,1),g)
  img.itemset((y,x,2),r)

  return img

for y in range(0, altura):
  for x in range(0, largura):

    #obj_img.itemset((y, x, 0), 0)
    #obj_img.itemset((y, x, 1), 0)

    azul, verde, vermelho = getColor(obj_img,x,y)
    setColor(obj_img, x, y, 0, verde, vermelho)

showImage(obj_img)
```
