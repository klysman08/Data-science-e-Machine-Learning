```
#matplotlib

import matplotlib.pyplot as plt

t =np.linspace(0, 2*np.pi, 100)
```

```
y1 = np.sin(t)
y2 = np.cos(t)

plt.figure()
plt.plot(t,y1, label='seno')
plt.plot(t,y2, label='cosseno')
plt.xlabel('$t$')
plt.grid()
plt.show()
```

```
import pandas as pd

data = pd.read_csv("https://s3.amazonaws.com/pycourse/births1880.txt",
                    names=['Names', 'Births'])
data.head(10)
                    
```

```
data.info()
```

```
data.describe()
```

```
#Agrupamento por nnome
data_name = data.groupby('Names')

#Aplicando soma

grouped_name = data_name.sum()
grouped_name
```

```
grouped_name.plot.bar(title='Nomes de bebes: ')
plt.show()
```

```
grouped_name.plot.pie(x='Names', y='Births', autopct='%1.1f%%', figsize=(8,8))
plt.show()
```

```
cars = pd.read_csv("https://s3.amazonaws.com/pycourse/cars.csv")
cars.head()
```

```
cars = cars.dropna()
```

```
cars.plot(x='wt', y='mpg', kind='scatter')
plt.xlabel ('Peso')
plt.ylabel ('Eficiencia em mpg')
plt.show()
```

```
cars_cyl = cars.groupby('cyl')
for name, data in cars_cyl:
  plt.plot(data['wt'].values, data['mpg'].values, 'o', label=str(name) + 'cly')
plt.xlabel('peso')
plt.ylabel('mpg')
plt.legend()
plt.show()
```

```
cars.boxplot(column='mpg', by='cyl')
plt.subtitle(' ')
plt.show()
```
