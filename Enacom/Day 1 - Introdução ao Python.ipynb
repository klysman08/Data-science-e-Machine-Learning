```
#Função

def sing(x):
  if x > 0:
    return 'positive'
  elif x < 0:
    return 'negative'
  else:
    return 'zero'

xl = [sing(x) for x in [-1,0,1]]
print (xl)
```

```
def hello(nome, loud=True):
  if loud == True:
    return print('Hello ' + nome.upper())
  if loud == False:
    return print('Hello ' + nome.capitalize())

hello('klysman', False)

hello(loud=False, nome='ENACOM')
  
```

```
#Classes

class Greeter:
  #construto
  def __init__(self, name):
    self.name = name
    
  #metodos
  def greet (self, loud=False):
    if loud:
      return print('Hello ' + self.name.capitalize())
    else:
      return print('Hello ' + self.name.upper())
    
```

```
g = Greeter('Klysman')
g.greet(loud=True)
```
