# Introdução a Matplotlib

## Importando a biblioteca
Note que existe uma nomenclatura convencional <font color=blue>plt</font> para facilitar o entendimento do seu código. <br>
O argumento <font color=blue>inline</font> permite que plots apareçam na linha onde foram chamados. Não sendo necessário chamar a função <font color=blue>show()</font>

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

## Função <font color=blue>plot()</font>

```python
x = np.arange(10)
y = np.arange(10)
```

```python
plt.plot(x, y)
# plt.show()  # Não é necessário por conta do comando na célula [2]
```

## Múltiplos plots, rotulação de eixos e título

```python
y2 = x*x
y2 = y2[::-1]
plt.plot(x, y, 'g^', x, y2, 'g-', x, y, 'r-')

plt.xlabel('eixo x')
plt.ylabel('eixo y')
plt.title('Meu grafico')
```

## Legenda

```python
plt.plot(x, y, 'g^', label='Primeiro')
plt.plot(x, y2, 'g-', label = 'Segundo') 
plt.plot(x, y, 'r-', label = 'Terceiro')

plt.xlabel('eixo x')
plt.ylabel('eixo y')
plt.title('Meu grafico')
plt.legend()
```

## Subplots

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(14,3), sharey=True)

ax1.plot(x, y, 'g^')
ax2.plot(x, y2, 'g-')
ax3.plot(x, y, 'r-')
```

## Múltiplos eixos na mesma figura

```python
fig = plt.figure()

ax1 = fig.add_axes([0.1, 0.1, 0.9, 0.9])
ax1.plot(x, y2, 'g-')

ax2 = fig.add_axes([0.6, 0.6, 0.3, 0.3])
ax2.plot(x, y, 'r-')
```

## Histogram

```python
mu, sigma = 0, 1
x = mu + sigma * np.random.randn(1000)

plt.hist(x, bins='auto')
```

## Heatmap

```python
from sklearn.metrics import confusion_matrix 
```

```python
y_true = [2, 0, 2, 2, 0, 1, 1, 1, 1]
y_pred = [0, 0, 2, 2, 0, 2, 1, 0, 1]

labels = np.unique(y_true)
label_names = ['A', 'B', 'C']

cm = confusion_matrix(y_true, y_pred)

plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.colorbar()

plt.xticks(labels, label_names)
plt.yticks(labels, label_names)
```

## Scatter

```python
num_samples = 300
class_A = [np.linspace(0,1,num_samples)-np.random.rand(num_samples), #X
         np.linspace(0,1,num_samples)-np.random.rand(num_samples)] #Y

class_B  = [np.linspace(1,2,num_samples)-np.random.rand(num_samples), #X
         np.linspace(0,1,num_samples)-np.random.rand(num_samples)] #Y

fig, ax = plt.subplots()
ax.scatter(class_A[0], class_A[1])
ax.scatter(class_B[0], class_B[1])
```

## Plot dinâmico 

```python
%matplotlib notebook
```

```python
import time
```

```python
np.linspace(1,0,50)
```

```python
fig, ax = plt.subplots() 
train, test = [], []

for i in np.linspace(1,0,50):
    ax.clear()
    
    train.append(i**3)
    test.append(i**3  + 0.2*np.random.rand() )
    
    ax.plot(train, 'r-')
    ax.plot(test, 'g-.')
    
    
    fig.canvas.draw()
    time.sleep(.2)
```
