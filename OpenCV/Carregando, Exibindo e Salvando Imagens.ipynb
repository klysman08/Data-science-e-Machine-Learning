```
import cv2
from google.colab.patches import cv2_imshow
```

Carregando uma imagem

```
imagemCarregada = cv2.imread("kl.jpg", 0)
```

Exibindo uma imagem no google colab

```
cv2_imshow(imagemCarregada)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Salvando a imagem

```
   cv2.imwrite("ImagemSalva.jpg", imagemCarregada)
```
