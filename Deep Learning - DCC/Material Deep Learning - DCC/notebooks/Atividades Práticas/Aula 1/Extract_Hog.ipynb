```python
import matplotlib.pyplot as plt

from skimage import io, feature, util
import os
import numpy as np

# Image sets.
img_set = ['train', 'test']

# Iterating over sets.
for curr_set in img_set:
    
    # Concat directories.
    img_dir = '../../../data/mnist/' + curr_set + '/'
    npy_dir = '../../../data/mnist_feats/' + curr_set + '/'

    # List Comprehension for files in a directory.
    files = [f for f in os.listdir(img_dir) if os.path.isfile(os.path.join(img_dir, f))]

    # Iterating over files.
    for it, f in enumerate(files):
        
        # Ignoring classes 2-9.
        if int(f[-5]) > 1:
            continue
        
        # Concat image paths.
        print(f + ' -> ' + f.replace('.png', '.npy'))
        
        img_path = os.path.join(img_dir, f)
        npy_path = os.path.join(npy_dir, f.replace('.png', '.npy'))
        
        # Reading image.
        img = io.imread(img_path)
        
        # Adding salt loss.
        img = util.random_noise(img, mode='salt', amount=0.3)
        
        # Extracting HOG features.
        hog_feat = feature.hog(img, orientations=8, pixels_per_cell=(7, 7), cells_per_block=(1, 1), multichannel=False)
        #print('hog', hog_feat.shape)
        
        # Saving npy file.
        np.save(npy_path, hog_feat)

print('End of files...')
```
