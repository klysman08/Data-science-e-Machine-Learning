```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
data = np.genfromtxt('iris.data', delimiter=',', usecols=(0,1,2,3))
len(data)
```

```
comprimentosetotas = data[:50,0]
```

```
plt.plot(comprimentosetotas, ls=':', marker='s', ms=6, label='C setota')
plt.plot(data[50:100,0], ls=':', marker='s', ms=6, label='C Versicolor') # comprimento das ires setosas
plt.legend()
plt.show()
```
