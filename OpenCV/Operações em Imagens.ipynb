```
import cv2
import numpy as np
```

Carregando imagem

```
imagem = cv2.imread("kl.jpg")
```

Obetendo informações da imagem
(dimenções e canais)

```
print (imagem.shape)
print (imagem.item(0, 0, 2),imagem.item(0, 0, 1),imagem.item(0, 0, 0))
```

Modificando cor de um pixel



```
imagem.itemset((0,0,2), 255)
imagem.itemset((0,0,1), 0)
imagem.itemset((0,0,0), 0)
```

Salvando a imagem modificada

```
cv2.imwrite("imagemSalva.jpg", imagem)
```

Extraindo região de interesse

```
regiao = imagem[1800:2000, 250:1015]
```

```
cv2.imwrite("RegiaoSalva.jpg", regiao)
```

Sobreponto imagens

```
imagem[2000:2200, 250:1015] = regiao
```

```
cv2.imwrite("Sobreposicao.jpg", imagem)
```
