```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = np.genfromtxt('iris.data', delimiter=',', usecols=(0,1,2,3))
len(data)
```

```python
comprimentosetotas = data[:50,0]
```

```python
plt.plot(comprimentosetotas, ls=':', marker='s', ms=9, label='C setotooa')
plt.plot(data[50:200,0], ls=':', marker='s', ms=9, label='C Versicolor') # comprimento das ires setosas
plt.legend()
plt.show()
```
