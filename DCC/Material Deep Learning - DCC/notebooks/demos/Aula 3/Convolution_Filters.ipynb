```python
import matplotlib.pyplot as plt

from skimage import io

from scipy import signal
import numpy as np

%matplotlib inline

arr = np.zeros((5, 5), dtype=np.float32)
arr[:, 0] = 10
arr[:, 1] = 15
arr[:, 2] = 20
arr[:, 3] = 80
arr[:, 4] = 85

arr = io.imread('emily.jpg')[:,:,0]

# Mean.
# kernel = np.zeros((3, 3), dtype=np.float32)
# kernel[:] = 1.0/9.0

# Mean 9x9.
# kernel = np.zeros((9, 9), dtype=np.float32)
# kernel[:] = 1.0/(9**2)


# Sobel horizontal.
# kernel = np.zeros((3, 3), dtype=np.float32)
# kernel[:, 0] = -1
# kernel[:, 2] = 1

# Sobel horizontal reverted.
# kernel = np.zeros((3, 3), dtype=np.float32)
# kernel[:, 0] = 1
# kernel[:, 2] = -1

# Sobel vertical.
# kernel = np.zeros((3, 3), dtype=np.float32)
# kernel[0, :] = -1
# kernel[2, :] = 1

# Laplace 3x3.
# kernel = np.ones((3, 3), dtype=np.float32) * -1
# kernel[1, 1] = 8

# Laplace 5x5.
# kernel = np.asarray([[0, 0,   1, 0, 0],
#                      [0, 1,   2, 1, 0],
#                      [1, 2, -16, 2, 1],
#                      [0, 1,   2, 1, 0],
#                      [0, 0,   1, 0, 0]], dtype=np.float) # Approximate 5x5 Laplacian.

result = signal.convolve2d(arr, kernel, 'same', boundary='fill')
# result = signal.convolve2d(arr, kernel, 'same', boundary='symm')

fig, ax = plt.subplots(1, 1, figsize=(10, 10))

print(arr)
ax.imshow(arr, cmap=plt.cm.gray)
ax.set_yticks([])
ax.set_xticks([])
ax.set_title('Image')

plt.show()

fig, ax = plt.subplots(2, 2, figsize=(20, 15))
print(kernel)
ax[0, 0].imshow(kernel, cmap=plt.cm.gray)
ax[0, 0].set_yticks([])
ax[0, 0].set_xticks([])
ax[0, 0].set_title('Kernel')

print(kernel[::-1, ::-1])
ax[0, 1].imshow(kernel[::-1, ::-1], cmap=plt.cm.gray)
ax[0, 1].set_yticks([])
ax[0, 1].set_xticks([])
ax[0, 1].set_title('Reverted Kernel')

print(result)
ax[1, 0].imshow(result, cmap=plt.cm.gray)
ax[1, 0].set_yticks([])
ax[1, 0].set_xticks([])
ax[1, 0].set_title('Result')

print(np.abs(result))
ax[1, 1].imshow(np.abs(result), cmap=plt.cm.gray)
ax[1, 1].set_yticks([])
ax[1, 1].set_xticks([])
ax[1, 1].set_title('Result ABS')

plt.show()
```
